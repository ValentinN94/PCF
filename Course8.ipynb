### Arbitrary arguments

- Use the *(unpacking operator) to create a arbitrary argument in a function
- Most common name is *args
- The unpacking operator will take all arguments given and unpack them into a tuple

```python
def my_sum_arb(*args):
    sum_args = 0
    for i in args:
        sum_args += i
    return sum_args
print(my_sum_arb(10, 25, 2.5, -50, 100, 30))
```

### Keyword arguments

- Use the **( keyword unpacking operator) to create a keyword argument in a function
- Most common name is **kwargs
- The unpacking operator will take the keys and the values given and unpack them into a dictionary

```python
def my_print_fun(**kwargs):
    for key, value in kwargs.items():
        print(key)
        print(value)
my_print_fun(name="John", lastname="Doe", age=30, dob='10.12.1990')
```

## Scoping

- A scope is a region in which your variable is recognized
- Also known as namespaces
- Scope in python follows the LEGB rule
- Local(L): Defined inside function/class
- Enclosed(E): Defined inside enclosing functions(Nested function concept)
- Global(G): Defined at the uppermost level
- Built-in(B): Reserved names in Python builtin modules

```python
# Global scope
x = 0

def outer():
    #Enclosed scope
    x = 1
    def inner():
        #Local scope
        x = 2
```

```python
i = 10 #In which scope are we now?
def modify_i():
    i = 20 #How about the scope here?
modify_i()
#What will be the value of i?
print(i)
```
